```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

df = pd.read_csv('11235/digest.csv')
alpha = set(df.alpha)

for a in sorted(alpha):
    df[df.alpha==a].loc[:, ['reached destination', 'mistakes', 'rewards', 'gamma']].plot(x='gamma', 
                                                                              subplots=True, 
                                                                              title='alpha: {}'.format(a))
    
df
```
